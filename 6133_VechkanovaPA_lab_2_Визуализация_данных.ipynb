#### Построить несколько графиков разного вида с использованием

```python
import matplotlib.pyplot as plt

# Гистограмма температуры
plt.figure(figsize=(6, 4))
plt.hist(stars['Temperature (K)'], bins=20, color='crimson', alpha=0.7)
plt.title('Распределение температуры')
plt.xlabel('Температура (K)')
plt.ylabel('Количество звёзд')
plt.grid(True)
plt.show()

print("")

# Линейный график зависимости радиуса от температуры
plt.figure(figsize=(6, 4))
plt.plot(stars['Temperature (K)'], stars['Radius(R/Ro)'], 'go', alpha=0.5)
plt.title('Радиус vs Температура')
plt.xlabel('Температура (K)')
plt.ylabel('Радиус (R/Ro)')
plt.gca().invert_xaxis()
plt.show()

print("")

# Boxplot абсолютной величины
plt.figure(figsize=(6, 4))
plt.boxplot(stars['Absolute magnitude(Mv)'], vert=False, patch_artist=True)
plt.title('Распределение абсолютной величины')
plt.xlabel('Absolute magnitude(Mv)')
plt.show()
```

#### Построить несколько суб-графиков на одном графике используя `.subplot`.  

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(11, 3))

# Гистограмма температуры
axes[0].hist(stars['Temperature (K)'], bins=20, color='teal', alpha=0.7)
axes[0].set_title('Распределение температуры')
axes[0].set_xlabel('Температура (K)')
axes[0].set_ylabel('Количество звёзд')
axes[0].grid(True)

# Точечный график (HR-диаграмма)
axes[1].scatter(
    stars['Temperature (K)'],
    stars['Luminosity(L/Lo)'],
    c=stars['Temperature (K)'],
    cmap='viridis',
    alpha=0.7
)
axes[1].set_title('HR-диаграмма')
axes[1].set_xlabel('Температура (K) →')
axes[1].set_ylabel('Светимость (L/Lo)')
axes[1].invert_xaxis()

# Столбчатая диаграмма
spectral_counts = stars['Spectral Class'].value_counts()
axes[2].bar(spectral_counts.index, spectral_counts.values, color='orange')
axes[2].set_title('Спектральные классы')
axes[2].set_xlabel('Спектральный класс')
axes[2].set_ylabel('Количество звёзд')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

#### Построить интерактивный график со слайдером.

```python
from ipywidgets import interact, IntRangeSlider

def update_hr_diagram(temp_range):
    """Обновление графика при изменении диапазона температур."""
    plt.figure(figsize=(8, 5))
    mask = (stars['Temperature (K)'] >= temp_range[0]) & (stars['Temperature (K)'] <= temp_range[1])
    filtered_data = stars[mask]
    plt.scatter(
        filtered_data['Temperature (K)'],
        filtered_data['Luminosity(L/Lo)'],
        c=filtered_data['Temperature (K)'],
        cmap='plasma',
        alpha=0.7
    )
    plt.title(f'HR-диаграмма: {temp_range[0]}K - {temp_range[1]}K')
    plt.xlabel('Температура (K) →')
    plt.ylabel('Светимость (L/Lo)')
    plt.gca().invert_xaxis()
    plt.colorbar(label='Температура (K)')
    plt.show()
    return

interact(
    update_hr_diagram,
    temp_range=IntRangeSlider(
        min=2000,
        max=40000,
        step=500,
        value=(3000, 20000),
        description='Диапазон температур:'
    )
)
```
