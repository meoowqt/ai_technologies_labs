### 1. Считать в `pandas.DataFrame` источник данных: CSV, JSON, Excel-файл, HTML-таблицу

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_covtype
```

```python
data = fetch_covtype()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Cover_Type'] = data.target
df
```

### 4. Продемонстрировать работу `.groupby`, на основе группировок в `groupby` вычислить агрегатные функции по одной или нескольким колонкам.

```python
grouped_cover = df.groupby('Cover_Type').agg({
    'Elevation': 'mean',
    'Slope': 'max'
}).rename(columns={
    'Elevation': 'Mean_Elevation',
    'Slope': 'Max_Slope'
})

grouped_cover
```

### 5. Решейпинг данных $1D\rightarrow2D$ с использованием `.pivot` (можно подать на вход результаты агрегатов, полученных ранее с использованием `.groupby` (сгруппировать по двум полям).

```python
grouped_for_pivot = df.groupby(['Cover_Type', 'Elevation_Category'])['Elevation'].mean().reset_index()

pivot_df = grouped_for_pivot.pivot(
    index='Cover_Type',
    columns='Elevation_Category',
    values='Elevation'
)

pivot_df
```

### 6. Решейпинг $1D\rightarrow2D$ данных, соединённых с группировкой / агрегацией (одним словом - сводная таблица): `.pivot_table`. Группировать только по категориальным полям или числовым, если значений немного.

```python
pivot_table_count = pd.pivot_table(
    df,
    index='Cover_Type',
    columns='Elevation_Category',
    values='Elevation',
    aggfunc='count',
    fill_value=0
)
pivot_table_count
```

### 7. Вычислить квантили распределения какого-либо вещественного признака (с использованием `numpy.quantile` или `numpy.percentile`).  

```python
quantiles = np.quantile(df['Elevation'], [0.25, 0.5, 0.75])

print(f"25%: {quantiles[0]:.2f}")
print(f"50%: {quantiles[1]:.2f}")
print(f"75%: {quantiles[2]:.2f}")
```

### 8. Вычислить (в виде текста) гистограмму какого-либо вещественного признака (с использованием `numpy.histogram`). Значения гистограммы можно использовать в качестве загрубленного числового признака для заданий (5) или (6).

```python
counts, bins = np.histogram(df['Elevation'], bins=5)

for i in range(len(counts)):
    bin_start = bins[i]
    bin_end = bins[i + 1]
    print(f"[{bin_start:.2f} – {bin_end:.2f}): {counts[i]} значений")
```

### 9. Проитерировать `DataFrame` построчно `.iterrows()` и выполнить какую-либо операцию внутри цикла.  

```python
count = 0
for _, row in df.iterrows():
    if row['Elevation'] > 3500:
        count += 1
print(f"\nКоличество строк с Elevation > 3500: {count}")
```

### 10. Получить `DataFrame` с `MultiIndex` любым способом: через конструктор (в документации есть множество видов конструкторов для создания `MultiIndex` с нуля), через `read_sql` / `read_csv` / `read_excel`, `read_*`, через `pivot_table`, через `groupby` или иными способами.  


```python
#Создаём MultiIndex через pivot_table
pivot_multi = pd.pivot_table(
    df,
    index=['Cover_Type', 'Elevation_Category'],
    values='Elevation',
    aggfunc='mean'
)
pivot_multi.head()
```

```python
# Меняем порядок уровней (Cover_Type и Elevation_Category):
pivot_multi_swapped = pivot_multi.swaplevel(0, 1)

pivot_multi_swapped.head()
```

```python
# Транспонируемтаблицу
transposed_multi = pivot_multi.T
transposed_multi.head()
```

```python
# Удаляем уровень:
pivot_dropped = pivot_multi.droplevel(level='Elevation_Category')
pivot_dropped.head()
```
